Film Özellikleri Veri Kümesi Analizi

```python
sbn.set(style="whitegrid")
plt.figure(figsize=(15, 25))
ax = sbn.barplot(y='Genre', x='Released_Year', data=newDF, palette='rocket')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Genre', fontsize=2)

# Y ekseninde yazıların daha net görünmesi için tick_params kullanalım
ax.tick_params(axis='y', which='major', pad=10)  # Y eksenindeki yazılar için pad değerini ayarlayabilirsiniz

plt.tight_layout()  # Yazıların sıkışmasını engellemek için

plt.show()
```
